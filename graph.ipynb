```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bereshitLanding.csv', encoding = 'ISO-8859-1', names=["time", "vs", "hs", 
"dist", "alt", "ang", "weight", "acc"])
df = df.drop([0])
df.head()
```

```python
df = df.astype(float)
df.info()
```

```python
df.plot(x="time", y="alt", kind="line", figsize=(10,10))
plt.show()
```

```python
df.plot(x="time", y="vs", kind="line", figsize=(10,10))
plt.show()
```

```python
df.plot(x="time", y="hs", kind="line", figsize=(10,10))
plt.show()
```
